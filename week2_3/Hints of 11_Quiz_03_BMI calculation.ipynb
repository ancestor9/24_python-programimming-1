## chat GPT에게 물어보면서 코딩하기

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Replace this URL with the actual URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Body_mass_index"

# Fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the BMI table by inspecting the Wikipedia page HTML
table = soup.find('table', class_='wikitable')

# Read the table data using pandas
df = pd.read_html(str(table))[0]
df
```

```python
df['Quiz_level'] = ['Under Weigth', 'Under Weigth', 'Under Weigth',
                    'Normal Weight', 'Over Weight', 'Obese',
                    'Clinicaly Obese', 'Clinicaly Obese']
df
```
