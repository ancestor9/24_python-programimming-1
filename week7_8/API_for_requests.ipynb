URL은 게시글 데이터를 제공하며, 각 게시글은 사용자 ID, 게시글 ID, 제목 및 본문을 포함합니다. 데이터는 JSON 형식으로 제공됩니다. JSONPlaceholder API는 무료이며, API 키 없이도 쉽게 사용할 수 있어 API 요청을 테스트하거나 프로토 타이핑하기에 좋은 리소스

```python
import requests

api_url = 'https://jsonplaceholder.typicode.com/posts'


def fetch_posts(api_url, max_pages=10):
    all_posts = []
    for page in range(1, max_pages + 1):
        response = requests.get(f"{api_url}?_page={page}")
        if response.status_code == 200:
            print('connection sucess')
            posts = response.json()
            all_posts.extend(posts)
            # Stop fetching if the page is empty
            if not posts:
                break
        else:
            print(f"Failed to retrieve data: Status code {response.status_code}")
            break
    return all_posts
```

```python
fetch_posts(api_url, max_pages=10)
```

```python
data = fetch_posts(api_url, max_pages=10)
```
