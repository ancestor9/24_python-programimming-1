```python
import socket
from datetime import datetime

# ngrok에서 생성한 URL을 사용하여 서버에 연결
ngrok_url = "tcp://8.tcp.ngrok.io:14451"  # Colab(Colab_server_ngrok.ipynb)에서 출력된 URL을 여기로 복사하세요
ngrok_host, ngrok_port = ngrok_url.replace("tcp://", "").split(":")
ngrok_port = int(ngrok_port)

address = (ngrok_host, ngrok_port)
max_size = 1000

print('Starting the client at', datetime.now())
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client.connect(address)
client.sendall(b'Hey!')
data = client.recv(max_size)
print('At', datetime.now(), 'someone replied', data)
client.close()
```
