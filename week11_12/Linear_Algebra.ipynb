# **Vectors**

```python
from typing import List

Vector = List[float]

def add(v: Vector, w: Vector) -> Vector:
    """Adds corresponding elements"""
    assert len(v) == len(w), "vectors must be the same length"

    return [v_i + w_i for v_i, w_i in zip(v, w)]

assert add([1, 2, 3], [4, 5, 6]) == [5, 7, 9]
```

```python
assert add([1, 2,], [2, 1]) == [3, 3]
```

### **dot, inner product**
- **The dot product of two vectors is the sum of their componentwise products**

```python
def my_dot(v, w):
    '''
    v_1 * w_1 + ... + v_n * w_n
    '''

    return sum([v_i * w_i for v_i, w_i in zip(v, w)])
```

### **Understanding inner product step by step**
> #### **내적은 두 벡터가 얼마나 같은 방향을 향하고 있는지를 측정**
> #### **벡터의 내적은 두 벡터 간의 각도와 관련있는 스칼라 값을 생성**

<img src = 'https://upload.wikimedia.org/wikipedia/commons/thumb/3/3e/Dot_Product.svg/330px-Dot_Product.svg.png'>


##### **이 그림 (A⋅B)B 는 벡터 A를 B 위에 투영한 결과**로 이는 A의 B방향 성분만을 추출하는 것과 같으며, 수학적으로는 다음과 같이 계산;
>> **투영(Projection)은 수직 그림자의 길이**
$$
\text{Proj}_{\mathbf{B}} \mathbf{A} = \frac{\mathbf{A} \cdot \mathbf{B}}{\mathbf{B} \cdot \mathbf{B}} \mathbf{B}
$$
>>

>> #### A와 B 내적을 A크기의 제곱으로 나눈 후, 이를 다시 B를 곱하면 결과적으로 B의 A에 대한 투영


```python
import numpy as np

# 벡터 A와 B 정의
A = np.array([3, 5])
B = np.array([6, 1])

# A와 B의 내적
dot_product = np.dot(A, B)
print(f"A와 B의 내적은 {dot_product}")

# B의 내적
B_dot_B = np.dot(B, B)

# 벡터 A의 벡터 B 위의 투영
proj_B_A = (dot_product / B_dot_B) * B

print("벡터 A의 벡터 B 위의 투영:", proj_B_A)
```

```python
dot_product/B_dot_B*B
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 벡터 A와 B 정의
A = np.array([3, 5])
B = np.array([6, 1])

# 벡터 A와 B의 내적 계산
dot_product = np.dot(A, B)

# B의 내적
B_dot_B = np.dot(B, B)

# 벡터 A의 벡터 B 위의 투영 계산
proj_B_A = (dot_product / B_dot_B) * B

# 그래프 그리기
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
plt.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')
plt.quiver(0, 0, proj_B_A[0], proj_B_A[1], angles='xy', scale_units='xy', scale=1, color='g', label='Proj_B A')

# 추가: Proj_B A 점선으로 표시
plt.plot([proj_B_A[0], A[0]], [proj_B_A[1], A[1]], 'g--')

# 축 설정
plt.xlim(-1, 8)
plt.ylim(-1, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# 벡터 레이블 추가
plt.text(A[0] * 0.5, A[1] * 0.5, 'A', color='red', fontsize=12)
plt.text(B[0] * 0.5, B[1] * 0.5, 'B', color='blue', fontsize=12)
plt.text(proj_B_A[0] * 0.5, proj_B_A[1] * 0.5, 'Proj_B A', color='green', fontsize=12)

# 범례 표시
plt.legend()

# 플롯 보여주기
plt.show()
```

```python
import requests
from PIL import Image
from io import BytesIO

# URL of the image
image_url = "https://people.com/thmb/h67yba4NaGbpnv9DZVZm894kSjo=/1500x0/filters:no_upscale():max_bytes(150000):strip_icc():focal(979x292:981x294)/Brad-Pitt_1-2000-e8a294b80b034e659785ad813d3f02f2.jpg"

# Send a GET request to the URL
response = requests.get(image_url)

# Check if the request was successful (status code 200)
print(response.status_code)

# Open the image using PIL
image = Image.open(BytesIO(response.content))
image.show()
```

```python
image
```

```python
image_array = np.array(image)
img = image_array[:,:,0]
plt.imshow(img)
```

```python
# Apply the shear transformation to the image

# Define the shear factor (adjust as needed)
shear_factor = 1.2

# Create a shear matrix for horizontal shear
shear_matrix = (1, shear_factor, 0, 0, 1, 0)

sheared_image = image.transform(image.size, Image.AFFINE, shear_matrix, Image.BICUBIC)
sheared_image
```
