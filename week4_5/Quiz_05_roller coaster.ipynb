## If-elif with 실습과제

````python
weight = int(input(print("몸무게(Kg)를 입력해주세요")))
height = int(input(print("신장(cm)를 입력해주세요")))
bmi = weight/(height/100)**2
````

````python
if bmi < 18.5:
    category = "Underweight"
    print(category)
if 18.5 <= bmi < 24.9:
    category = "Normal weight"
    print(category)
if 25 <= bmi < 29.9:
    category = "Overweight"
    print(category)
if bmi >= 30:
    category = "Obese"
    print(category)
````
