### **실습과제 1 : 우회전으로 다시 제자리에 돌아오는 프로그램을 만들어라**
- **단 For loop를 사용하라**

```python
clearscreen()
shape("turtle")
setup(300,300)
for i in range(4):
    forward(100)
    right(90)
```

```python
for _ in range(4):
    print('a')
```

```python
clearscreen()
shape("turtle")
for _ in range(4):
    forward(100)
    right(90)

left(90)
for _ in range(4):
    forward(100)
    left(90)
```

```python
iNumber = 23

for i in range(1000000):
    iGuess = int(input('Enter an integer: '))
    if iGuess == iNumber:
        print('Congratulations, you guessed it.')
        break
    elif iGuess < iNumber:
        print('No, it is a little higher than that.')
    else:
        print('No, it is a little lower than that.')
```
